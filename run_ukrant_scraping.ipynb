```python
from web_scraping.scrape import scrape_ukrant_category
import json
```

```python
news_headlines_dutch = scrape_ukrant_category(
    url="https://ukrant.nl/categorie/nieuws-nl/",  # scraping the Dutch news category
    category="nieuws"
)
```

```python
news_headlines = scrape_ukrant_category(
    url="https://ukrant.nl/categorie/nieuws-en/?lang=en", 
    category="news"
)
```

```python
background_headlines = scrape_ukrant_category(
    url="https://ukrant.nl/categorie/background-en/?lang=en", 
    category="background"
)
```

```python
science_headlines = scrape_ukrant_category(
    url="https://ukrant.nl/categorie/science-en/?lang=en", 
    category="science"
)
```

```python
# Combine the lists of articles
all_articles = news_headlines + background_headlines + science_headlines

# Save to a json file
with open("data/ukrant_headlines.json", "w", encoding="utf-8") as f:
    json.dump(all_articles, f, ensure_ascii=False, indent=2)

print(f"Saved {len(all_articles)} total headlines")
```
